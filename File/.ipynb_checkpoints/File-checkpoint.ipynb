## 파일 제어
* ```(파일명) = open('파일명', (액션), encoding='utf8')``` 형식으로 열 수 있다. 
* 액션 종류
 * 'r' -> 읽기 전용. 이미 생성된 파일만.
 * 'w' -> 쓰기, 파일을 생성하며, 기존의 파일은 덮어쓴다.
 * 'x' -> 쓰기, 새 파일에만 적용가능.
 * 'a' -> 쓰기, 파일을 생성하며, 기존의 파일의 뒷부분에 추가
 * '+' -> 읽기쓰기 모두
 * 'rb', 'wb' -> 바이너리 형식으로 파일 작성.  
 * 참조 : [파이썬 파일 사용법 정리](https://withcoding.com/86)
* 파일 읽기 함수
 * f.read() -> 파일 전체를 문자열로 읽는다. 
 * f.readline() -> 파일의 줄을 읽는다. 반복 실행시 
 * f.readlines() -> 각 줄을 문자열의 배열로 반환한다.
* 파일 쓰기 함수
 * f.write('문자열') -> 파일을 문자열로 작성한다.
 * f.writelines
* 파일을 읽고 쓰는 위치 찾기 
 * f.fileno() -> 파일의 번호를 확인하는 함수. 4번부터 시작
 * f.tell() -> 파일의 현재 위치를 반환한다.
 * f.seek(번호) -> 파일의 제어 위키를 지정한다.
* 파일 닫기 
 * f.close() 사용

````python
f = open('test.txt', 'r')
print(dir(f))
print(type(f))
f.close()
````

````python
# 예시 - 파일 읽기
print('read로 읽기')
f = open('test.txt', 'r', encoding='utf8')
# read로 읽기
print(f.fileno())
txt0 = f.read()
print(txt0)
f.close()
````

````python
# readlines로 읽기
print('readlines로 읽기')
f = open('test.txt', 'r', encoding='utf8')
txt_list = f.readlines()
print(type(txt_list))
print(txt_list)
for txt1 in txt_list:
    print(txt1)
f.close()
````

````python
# readline으로 읽기 - 한 줄만 읽는다.
print('readline으로 읽기')
f = open('test.txt', 'r', encoding='utf8')
txt1 = f.readline()
print(txt1, end="")
txt1 = f.readline()
print(txt1, end="")
f.close()
````

````python
f = open('test.txt', 'r', encoding='utf8')
s = f.readline()
print(s)
for s in f:
    print(s)
f.close()
````

````python
#tell, seek -> 위치 찾기, 위치 옮기기
f = open('test.txt', 'r', encoding='utf8')
txt1 = f.readline()
print(txt1, end="")
print('위치', f.tell())
txt2 = f.readline()
print(txt2, end="")
print(f.tell())
f.close()

g= open('LoremIpsum.txt', 'r')
print(g.fileno())
txt1 = g.readline()
print(txt1, end="")
print(g.tell())
print(g.fileno())
remtext = g.read()
print(remtext)
print()
print('다시 처음으로')
g.seek(0)
txt3 = g.readline()
print(txt3)
g.close()
````

````python
#파일 쓰기 - w나 a 옵션 사용 a는 이미 있는 파일 사용.
# 함수는 write나 writelines 사용.
h = open('new.txt', 'w+')
j = open('new/new1.txt', 'w+')
print(h.fileno(), j.fileno())
h.write('new Test!\n')
j.write('new Test!\n')
h.writelines(['attend', ':', 'LeeYoHan'])
j.writelines(['attend', ':', 'LeeYoHan'])
h.close()
j.close()
````

````python
h = open('new.txt', 'w')
j = open('new/new1.txt', 'a')
print(h.fileno(), j.fileno())
h.write('Overwrite!\n')
j.write('Add Text!\n')
h.writelines(['attend', ':', 'LeeYoHan'])
j.writelines(['attend', ':', 'LeeYoHan'])
h.close()
j.close()
````

### with문 
* 자동으로 파일 닫을 때 사용.
>```
>with open('txt파일', '조건') as (변수명):
>    (작업명)
>```

````python
with open ('test.txt', 'r', encoding='utf8') as f:
    print(f.read())
    print(f.tell())
````
